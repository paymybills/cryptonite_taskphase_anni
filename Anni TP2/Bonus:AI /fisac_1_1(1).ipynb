```python
deli.rename(columns={'match_id': 'id'}, inplace=True)
print(f"{deli.columns}, \n \n \n {match.columns}")
```

```python
#merge
merged_df = deli.merge(match, left_on='id', right_on='id', suffixes=('_delivery', '_match'))

total_runs_per_season = merged_df.groupby('season')['total_runs'].sum().reset_index()

total_runs_per_season.columns = ['season', 'total_runs']

print(total_runs_per_season)
```

```python
merged_df.head(15)
```

```python
# Group by batsman and sum the runs
top_scorers = merged_df.groupby('batsman')['batsman_runs'].sum().reset_index()

# Rename columns for clarity
top_scorers.columns = ['batsman', 'total_runs']

# Sort the DataFrame to get the top 10 run scorers
top_10_scorers = top_scorers.sort_values(by='total_runs', ascending=False).head(10)

# Display the top 10 scorers
print(top_10_scorers)
```

```python
sns.set(style="whitegrid")

# Create a bar plot for the top 10 scorers
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='batsman', y='total_runs', data=top_10_scorers, palette='viridis')

# Add a best fit line
x = np.arange(len(top_10_scorers))
y = top_10_scorers['total_runs']
z = np.polyfit(x, y, 1)  # Linear fit
p = np.poly1d(z)
plt.plot(x, p(x), color='red', linestyle='--')

# Customize the plot
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#Q22 total matches
all_teams = pd.concat([match['team1'], match['team2']])

matches_played = all_teams.value_counts()

print(matches_played)
```

```python
# Calculate total matches played by each team
total_matches = matches_played.to_frame(name='Total Matches')

# Calculate total wins for each team
wins = match['winner'].value_counts().to_frame(name='Wins')

# Merge total matches and wins data
team_stats = total_matches.merge(wins, left_index=True, right_index=True, how='left')
team_stats['Wins'].fillna(0, inplace=True)  # Fill NaN values with 0 for teams with no wins

# Calculate win rate
team_stats['Win Rate'] = (team_stats['Wins'] / team_stats['Total Matches']) * 100

# Display the results
print(team_stats)
```

```python
# Calculate total matches played by each team
total_matches = matches_played.to_frame(name='Total Matches')

# Calculate total wins for each team
wins = match['winner'].value_counts().to_frame(name='Wins')

# Merge total matches and wins data
team_stats = total_matches.merge(wins, left_index=True, right_index=True, how='left')
team_stats['Wins'].fillna(0, inplace=True)  # Fill NaN values with 0 for teams with no wins

# Calculate win rate
team_stats['Win Rate'] = (team_stats['Wins'] / team_stats['Total Matches']) * 100

# Display the results
print(team_stats)
```

```python
# Get the unique team names
teams = match['winner'].unique()

# Create a mapping from team names to numerical indices
team_mapping = {team: i for i, team in enumerate(teams)}

# Convert the 'winner' column to numerical indices
match['winner_numeric'] = match['winner'].map(team_mapping)

# Create a histogram of win distribution using the numerical indices
plt.figure(figsize=(10, 6))
plt.hist(match['winner_numeric'], bins=len(teams), edgecolor='black')

# Set x-axis ticks to team names
plt.xticks(np.arange(len(teams)) + 0.5, teams, rotation=45, ha='right')

plt.title('Distribution of Won Matches')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.tight_layout()
plt.show()
```

```python
# Filter for matches where the toss winner is the same as the match winner
toss_win_match_win = match[match['toss_winner'] == match['winner']]

# Group by toss winner and toss decision to get the counts
toss_decisions = toss_win_match_win.groupby(['toss_winner', 'toss_decision']).size().reset_index(name='count')

# Display the results
print(toss_decisions)
```

```python
# Create a bar plot using seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='toss_winner', y='count', hue='toss_decision', data=toss_decisions, palette='viridis')
plt.title('Toss Decisions by Each Team After Winning')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
